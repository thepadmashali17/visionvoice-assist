```python
import cv2
import torch
from gtts import gTTS
import pygame
import io
from collections import Counter
import tkinter as tk
from PIL import Image, ImageTk

# Initialize pygame
pygame.init()

# Load YOLOv5 model
model = torch.hub.load('ultralytics/yolov5', 'yolov5x', pretrained=True)

# Initialize webcam
cap = cv2.VideoCapture(0)

# Variable to store detected object labels
detected_objects = []

# Previous detected objects to prevent repeated announcements
prev_detected_objects = []

# Confidence threshold for detections
confidence_threshold = 0.6  # Adjust as needed

def speak(text):
    # Create a file-like object in memory
    memory_file = io.BytesIO()

    # Convert text to speech and save to memory file
    tts = gTTS(text=text, lang='en')
    tts.write_to_fp(memory_file)
    
    # Play the audio from memory file
    memory_file.seek(0)
    pygame.mixer.music.load(memory_file)
    pygame.mixer.music.play()

    # Wait until the audio finishes playing
    while pygame.mixer.music.get_busy():
        pygame.time.Clock().tick(10)

# Create tkinter window
window = tk.Tk()
window.title("VisionVoice Assist")

# Create a label for the heading
heading_label = tk.Label(window, text="VisionVoice Assist", font=("Helvetica", 16))
heading_label.pack()

# Create a label to display the object detection screen
label_image = tk.Label(window)
label_image.pack()

# Create a label to display the detected objects
detected_objects_label = tk.Label(window, text="Detected Objects:", font=("Helvetica", 12))
detected_objects_label.pack()

# Function to update the detected objects label
def update_detected_objects_label(detected_objects):
    detected_objects_label.config(text="Detected Objects: " + ', '.join(detected_objects))

def on_closing():
    # Release resources
    cap.release()
    cv2.destroyAllWindows()
    window.destroy()

# Bind the window closing event to on_closing function
window.protocol("WM_DELETE_WINDOW", on_closing)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Perform object detection
    results = model(frame)

    # Extract detected object labels with confidence above the threshold
    detected_objects = [model.names[int(cls)] for cls, conf in zip(results.pred[0][:, -1], results.pred[0][:, 4]) if conf >= confidence_threshold]

    # Count the occurrences of each detected object
    object_counts = Counter(detected_objects)

    # Generate text for detected objects
    text_parts = []
    for obj, count in object_counts.items():
        if count == 1:
            text_parts.append(f"a {obj}")
        else:
            text_parts.append(f"{count} {obj}s")
        
    # Convert detected object names to text
    text = "I see " + 'and '.join(text_parts) if text_parts else ""

    # Check if the detected objects have changed or not
    if prev_detected_objects != detected_objects:
        if text != "":
            # Convert text to speech
            speak(text)

        # Update previous detected objects
        prev_detected_objects = detected_objects.copy()

        # Update detected objects label
        update_detected_objects_label(detected_objects)

    # Apply color correction to the frame
    frame_rgb = cv2.cvtColor(results.render()[0], cv2.COLOR_BGR2RGB)

    # Render results
    img = Image.fromarray(frame_rgb)
    img = ImageTk.PhotoImage(image=img)
    label_image.imgtk = img
    label_image.config(image=img)

    # Update tkinter window
    window.update()

    # Check for key press
    key = cv2.waitKey(1)
    if key & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
```
